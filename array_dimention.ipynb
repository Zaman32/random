**1D Array concatenation**

```python
# 1D array

one_dim1 = np.arange(0, 5).reshape(5)
one_dim2 = np.arange(5, 10).reshape(5)
```

```python
print('one_dim1 ', one_dim1)
print('one_dim2 ', one_dim2)
```

```python
np.concatenate((one_dim1, one_dim2), axis=0)
```

**2D array concatenation**

```python
# 2D array

two_dim1 = np.arange(0,6).reshape(2,3)
two_dim2 = np.arange(6, 12).reshape(2,3)
print('two_dim1 = \n', two_dim1)
print('\n two_dim2 = \n', two_dim2)
```

```python
# concatenate over axis 0
two_dim_axis0 = np.concatenate((two_dim1, two_dim2), axis = 0)
```

```python
print('two_dim_axis0 = \n', two_dim_axis0)
print('\n shape = ', two_dim_axis0.shape)
```

```python
# concatenate over axis 1
two_dim_axis1 = np.concatenate((two_dim1, two_dim2), axis = 1)
```

```python
print('two_dim_axis1 = \n', two_dim_axis1)
print('\n shape = ', two_dim_axis1.shape)
```

**3D array concatenation**

```python
# 3D array
three_dim1 = np.arange(0, 24).reshape(2,3,4)
three_dim2 = np.arange(24, 48).reshape(2,3,4)

print(' \n three_dim1 = \n', three_dim1)
print(' \n three_dim2 = \n', three_dim2)
```

```python
# concatenate over axis 0
three_dim_axis0 = np.concatenate((three_dim1, three_dim2), axis=0)
```

```python
print('three_dim_axis0 = \n', three_dim_axis0)
print('\n shape = ',three_dim_axis0.shape)
```

```python
# concatenate over axis 1
three_dim_axis1 = np.concatenate((three_dim1, three_dim2), axis=1)
```

```python
print('three_dim_axis1 = \n', three_dim_axis1)
print('\n shape = ',three_dim_axis1.shape)
```

```python
# concatenate over axis 2
three_dim_axis2 = np.concatenate((three_dim1, three_dim2), axis=2)
```

```python
print('three_dim_axis2 = \n', three_dim_axis2)
print('\n shape = ', three_dim_axis2.shape)
```
